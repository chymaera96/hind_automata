```
def translate_notation(swaras):
  hindi = ['S',r'r','R','g','G','M','m','P','d','D','n','N']
  english = ['C','C#','D','D#','E','F','F#','G','G#','A','A#','B']
  trans =  dict(zip(hindi, english))
  note = []
  for s in swaras:
    if s!=',':
      octave = s[1]
      note.append(trans[s[0]]+octave)
    else:
      note.append(s)
  return note

def create_pairs(raga):
  bi = []
  for note1 in raga:
    for note2 in raga:
      if (note1,note2) not in bi:
        bi.append((note1,note2))
  return bi
```

```
# desh = ['S3','R3','G3','M3','P3','D3','n3','N3','S4','R4','G4','M4','P4','D4','n4','N4','S5','R5','G5','M5','P5','D5','n5','N5',',']
yaman = ['S3','R3','G3','m3','P3','D3','N3','S4','R4','G4','m4','P4','D4','N4','S5','R5','G5','m5','P5','D5','N5',',']
filename = '/content/yaman.dat'
unotes = translate_notation(yaman)
piecewise_transition_probs1 = []
p_transition_probs = np.ones((len(unotes),len(unotes)))
with open(filename) as fp:
  for line in fp:
    if len(line) > 0:
      line = line.strip()
      # Counting note transitions
      prev_note = ',' # start with a blank note
      print(line)
      notes = translate_notation(line.split(' '))
      for p_note in notes:
        if len(p_note) > 0:
          try:
            toInd = unotes.index(p_note)
            fromInd = unotes.index(prev_note)
            p_transition_probs[fromInd,toInd] += 1
          except:
            print('\nNote incompatible with raga found:'+p_note+", ignoring note")
          if p_note == ',': pass # retain previous note on sustain
          else: 
            prev_note = p_note # update previous note to current note if it was not sustain
```

```
# convert frequencies to probabilities (unigram)
ps = np.sum(p_transition_probs,1)
ps[np.where(ps==0)] = .1
p_transition_probs = np.transpose(np.transpose(p_transition_probs)/ps) 
piecewise_transition_probs1.append(p_transition_probs)

# transition probabilities matrix between the unique notes in the piece
transition_probs1 = np.zeros((len(unotes),len(unotes)))
# computing generic transition_probs from piecewise_transition_probs
for transition_prob in piecewise_transition_probs1:
    transition_probs1 += transition_prob
```

```
# convert frequencies to probabilities (bigram)
qs = np.sum(q_transition_probs,1)
qs[np.where(qs==0)] = .1
q_transition_probs = np.transpose(np.transpose(q_transition_probs)/qs) 
piecewise_transition_probs2.append(q_transition_probs)

# transition probabilities matrix between the unique notes in the piece
transition_probs2 = np.zeros((len(binotes),len(binotes)))
# computing generic transition_probs from piecewise_transition_probs
for transition_prob in piecewise_transition_probs2:
    transition_probs2 += transition_prob
```

```
# Simulation

note_space_duration = 0.75
num_beat_cycle = 25
alankar_density = 2
prev_swar_ind = int((len(binotes)-1)/3)  # Initialize the automata with the tonic in the middle octave
taal = 8   # Number of beats in a beat cycle. Choosing 'teental'
part = music21.stream.Part(id='sitar')
part.append(music21.instrument.Sitar())
prev_note = music21.note.Note(binotes[prev_swar_ind][0])
backup_swar_ind = prev_swar_ind
```

```
for cycle in range(num_beat_cycle):
  measure = music21.stream.Measure(number=1)
  notes_list = []
  if cycle == 0:
    print('k')
    note = music21.note.Note(binotes[backup_swar_ind][0])
    note.duration.quarterLength = note_space_duration
    print(note)
  else:
    for taal_num in range(taal):
      swar_ind = np.random.choice(np.arange(len(binotes)),p=transition_probs2[prev_swar_ind,:])
      alankar_ind = unotes.index(binotes[swar_ind][0])

      # Random seed for the alankar
      seed = random.uniform(-1, 1)
      if seed < 0 and taal_num != 0:
        notes_list[-1].duration.quarterLength = note_space_duration/alankar_density
        alankar = np.arange(note_space_duration/alankar_density,note_space_duration,note_space_duration/alankar_density)
        for ix in alankar:
          print("aaaaaa")
          if unotes[alankar_ind] == ',':
            notes_list[-1].duration.quarterLength += note_space_duration/alankar_density
          else:
            # insert a note with duration alankar_density
            this_note = music21.note.Note(unotes[alankar_ind])
            print(unotes[alankar_ind])
            this_note.duration.quarterLength = note_space_duration/alankar_density
            notes_list.append(this_note)

            # update previous swar to fill in next note in alankar
            swar_choice_ind = np.random.choice(np.arange(len(unotes)),p=transition_probs1[alankar_ind,:])
            swar_choice = unotes[swar_choice_ind]
            prev_swar_ind = binotes.index((unotes[alankar_ind],swar_choice))
            prev_note = this_note

            # get index for next note in alankar
            if seed < 0.5:
              alankar_ind += 1
            else:
              alankar_ind -= 1

            if alankar_ind > len(unotes): 
              alankar_ind -= 2
            if alankar_ind < 0: 
              alankar_ind += 2
      
      else:
        if binotes[swar_ind][1] == ',' and taal_num==0:
          # print("Wrong starting note!")
          while binotes[swar_ind][1] == ',':
              swar_ind = np.random.choice(np.arange(len(binotes)),p=transition_probs2[backup_swar_ind,:])

      # extend note duration if current index denotes a ','
      if binotes[swar_ind][1] == ',':
        notes_list[-1].duration.quarterLength += note_space_duration
      else:
        # create a note object for one quarterLength
        this_note = music21.note.Note(binotes[swar_ind][1])
        this_note.duration.quarterLength = note_space_duration
        # add it to the list
        notes_list.append(this_note)

        # update note and swar index for next iteration
        backup_swar_ind = prev_swar_ind
        prev_swar_ind = swar_ind
        prev_note = this_note


  for note in notes_list:
    print(note)
    measure.append(note)

  # add measure to part
  part.append(measure)
```

```
# create an empty score
simulation = music21.stream.Score()

# add the part to it
simulation.append(part)
fp = simulation.write('midi', fp='/content/yaman.mid')

```
